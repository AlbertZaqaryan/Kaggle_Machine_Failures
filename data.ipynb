```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train
```

```python
df_train.info()
```

```python
df_train['Type'].unique()
```

```python
df_train['Type'] = df_train['Type'].replace(to_replace='L', value=0)
df_train['Type'] = df_train['Type'].replace(to_replace='M', value=1)
df_train['Type'] = df_train['Type'].replace(to_replace='H', value=2)
```

```python
# df_train = df_train.drop('id', axis=1)
df_train = df_train.drop('Product ID', axis=1)
```

```python
df_train
```

```python
df_train.corr()['Machine failure']
```

```python
X_train = df_train.drop('Machine failure', axis=1)
y_train = df_train['Machine failure']
```

```python
X_train
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
model = RandomForestClassifier()
scaler = StandardScaler()
pipe = Pipeline([('scaler', scaler), ('rfc', model)])
```

```python
parameters = {
    'rfc__n_estimators':list(range(70, 73)),
    'rfc__criterion':["gini", "entropy", "log_loss"],
    'rfc__max_depth':[4, 5, 6],
    'rfc__max_features':["sqrt", "log2", None]
}
```

```python
full_model = GridSearchCV(pipe, param_grid=parameters, verbose=2, scoring='accuracy', cv=3)
```

```python
full_model.fit(X_train, y_train)
```

```python
full_model.best_params_
```

```python
y_pred = full_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```
